```python
#Q1

#We’ll first clean the columns AskPrice and kmDriven and handle missing data.
import pandas as pd

# Load the dataset
file_path = '/content/used_car_dataset.csv'
data = pd.read_csv(file_path)

# Display an overview of the dataset
data_overview = {
    "Columns": data.columns.tolist(),
    "Number of Rows": data.shape[0],
    "Number of Columns": data.shape[1],
    "Sample Data": data.head().to_dict(),
    "Data Types": data.dtypes.to_dict(),
}

data_overview
```

```python
#Let’s group the data by Brand to calculate the average AskPrice and visualize it with a bar chart.

# Clean the 'AskPrice' column: remove ₹ symbol and commas, then convert to numeric
data['AskPrice'] = data['AskPrice'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
data['AskPrice'] = pd.to_numeric(data['AskPrice'], errors='coerce')

# Clean the 'kmDriven' column: remove 'km' and commas, then convert to numeric
data['kmDriven'] = data['kmDriven'].astype(str).str.replace(' km', '', regex=False).str.replace(',', '', regex=False)
data['kmDriven'] = pd.to_numeric(data['kmDriven'], errors='coerce')

# Drop rows with missing or invalid 'AskPrice' or 'kmDriven'
data = data.dropna(subset=['AskPrice', 'kmDriven'])

# Display cleaned dataset summary
cleaned_data_summary = {
    "Number of Rows (After Cleaning)": data.shape[0],
    "Number of Columns": data.shape[1],
    "Sample Cleaned Data": data.head().to_dict(),
    "Data Types": data.dtypes.to_dict()
}
cleaned_data_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average asking price for each car brand
avg_price_by_brand = data.groupby('Brand')['AskPrice'].mean().sort_values(ascending=False)

# Plot the average asking price for the top 10 brands
plt.figure(figsize=(12, 6))
avg_price_by_brand.head(10).plot(kind='bar', color='skyblue')
plt.title('Average Asking Price by Brand (Top 10)')
plt.xlabel('Brand')
plt.ylabel('Average Asking Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q2

#Let’s analyze this by creating a boxplot to compare AskPrice across FuelType.

# Boxplot to compare asking price across fuel types
plt.figure(figsize=(10, 6))
sns.boxplot(x='FuelType', y='AskPrice', data=data)
plt.title('Asking Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Asking Price')
plt.show()
```

```python
#Q3

#We’ll use a scatter plot to analyze the relationship between Age and AskPrice and calculate the correlation coefficient.

# Scatter plot for Age vs. Asking Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='AskPrice', data=data, alpha=0.6)
plt.title('Car Age vs. Asking Price')
plt.xlabel('Car Age (years)')
plt.ylabel('Asking Price')
plt.show()

# Calculate the correlation coefficient
from scipy.stats import pearsonr

correlation, p_value = pearsonr(data['Age'], data['AskPrice'])
print(f"Correlation between Age and Asking Price: {correlation:.2f} (p-value: {p_value:.2e})")
```

```python
#Q4

#We’ll calculate the mean AskPrice for each Transmission type and compare using a bar chart.

# Calculate average asking price by transmission type

avg_price_by_transmission = data.groupby('Transmission')['AskPrice'].mean()

# Plot average price by transmission
plt.figure(figsize=(8, 6))
avg_price_by_transmission.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Average Asking Price by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Average Asking Price')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q5

#We’ll analyze this with a scatter plot and calculate the correlation coefficient.

# Scatter plot for Kilometers Driven vs. Asking Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='kmDriven', y='AskPrice', data=data, alpha=0.6)
plt.title('Kilometers Driven vs. Asking Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Asking Price')
plt.show()

# Calculate the correlation coefficient
correlation_km, p_value_km = pearsonr(data['kmDriven'], data['AskPrice'])
print(f"Correlation between Kilometers Driven and Asking Price: {correlation_km:.2f} (p-value: {p_value_km:.2e})")
```

```python
#Q6

#We’ll group by model, calculate the mean AskPrice, and visualize the top 5.

# Top 5 car models with the highest average asking price
avg_price_by_model = data.groupby('model')['AskPrice'].mean().sort_values(ascending=False).head(5)

# Plot top 5 models
plt.figure(figsize=(10, 6))
avg_price_by_model.plot(kind='bar', color='lightgreen')
plt.title('Top 5 Car Models with Highest Average Asking Price')
plt.xlabel('Car Model')
plt.ylabel('Average Asking Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q7

#We’ll analyze AskPrice distribution across ownership types using a boxplot.

# Boxplot for ownership history vs. asking price
plt.figure(figsize=(10, 6))
sns.boxplot(x='Owner', y='AskPrice', data=data)
plt.title('Asking Price by Ownership History')
plt.xlabel('Ownership History')
plt.ylabel('Asking Price')
plt.show()
```

```python
#Q8

#We’ll filter for the top 4 brands and create a countplot for FuelType.

# Top 4 most common car brands
top_brands = data['Brand'].value_counts().nlargest(4).index

# Filter data for top brands
filtered_data = data[data['Brand'].isin(top_brands)]

# Countplot for fuel types within top 4 brands
plt.figure(figsize=(12, 6))
sns.countplot(x='Brand', hue='FuelType', data=filtered_data)
plt.title('Fuel Types for Top 4 Car Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.legend(title='Fuel Type')
plt.show()
```

```python
#Q9

#We’ll use a KDE plot to compare price distributions for manual vs. automatic transmissions.

# KDE plot for price distribution by transmission type
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data[data['Transmission'] == 'Manual']['AskPrice'], label='Manual', shade=True)
sns.kdeplot(data=data[data['Transmission'] == 'Automatic']['AskPrice'], label='Automatic', shade=True)
plt.title('Asking Price Distribution by Transmission Type')
plt.xlabel('Asking Price')
plt.ylabel('Density')
plt.legend(title='Transmission')
plt.show()
```

```python
#Q10

#We’ll create a new column PricePerKm and compare average values across brands.

# Create a new column for price per kilometer
data['PricePerKm'] = data['AskPrice'] / data['kmDriven']

# Average price per kilometer by brand
avg_price_per_km = data.groupby('Brand')['PricePerKm'].mean().sort_values(ascending=True).head(10)

# Plot price per kilometer for top 10 brands
plt.figure(figsize=(12, 6))
avg_price_per_km.plot(kind='bar', color='purple')
plt.title('Price Per Kilometer by Brand (Top 10)')
plt.xlabel('Brand')
plt.ylabel('Average Price Per Kilometer')
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. Calculate the average price of cars by brand using NumPy.
# Find the mean asking price for each brand and visualize the top brands with a bar chart.

# Calculate average price by brand using NumPy
brands = data['Brand'].unique()
avg_price_by_brand_numpy = {brand: np.mean(data[data['Brand'] == brand]['AskPrice']) for brand in brands}

# Convert to a sorted DataFrame for visualization
avg_price_df = pd.DataFrame(avg_price_by_brand_numpy.items(), columns=['Brand', 'Average Price']).sort_values(by='Average Price', ascending=False)

# Visualize
plt.figure(figsize=(12, 6))
sns.barplot(x='Brand', y='Average Price', data=avg_price_df.head(10))
plt.title('Top 10 Brands by Average Asking Price')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

```

```python
# 12. Compute the variance and standard deviation of car prices by fuel type.
# Use NumPy to understand the spread of prices for each fuel type.

# Calculate variance and standard deviation using NumPy
fuel_types = data['FuelType'].unique()
var_std_by_fuel = {
    fuel: {
        "Variance": np.var(data[data['FuelType'] == fuel]['AskPrice']),
        "Standard Deviation": np.std(data[data['FuelType'] == fuel]['AskPrice'])
    } for fuel in fuel_types
}

# Display
pd.DataFrame(var_std_by_fuel).T
```

```python
# 13. Determine the proportion of manual vs. automatic cars.
# Calculate percentages for each transmission type and visualize with a pie chart.

# Calculate proportions
transmission_counts = data['Transmission'].value_counts(normalize=True)

# Pie chart
plt.figure(figsize=(8, 8))
transmission_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Manual vs. Automatic Cars')
plt.ylabel('')
plt.show()
```

```python
# 14. Categorize cars into different price ranges (budget, mid-range, luxury).
# Use NumPy's histogram to divide prices into bins and create a bar chart.

# Define price bins
price_bins = [0, 200000, 500000, 1000000, 2000000]
price_labels = ['Budget', 'Mid-Range', 'Luxury', 'Ultra Luxury']

# Create a new column for price range
data['PriceRange'] = pd.cut(data['AskPrice'], bins=price_bins, labels=price_labels)

# Countplot for price ranges
plt.figure(figsize=(10, 6))
sns.countplot(x='PriceRange', data=data, palette='viridis')
plt.title('Distribution of Cars by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

```python
# 15. Analyze the correlation between car age, kilometers driven, and asking price.
# Calculate a correlation matrix and visualize it with a heatmap.

# Calculate correlation matrix
correlation_matrix = data[['Age', 'kmDriven', 'AskPrice']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 16. Compare price distributions for the oldest (>20 years) and newest (<5 years) cars.
# Use boxplots to show differences in asking prices for old vs. new cars.

# Filter oldest and newest cars
old_cars = data[data['Age'] > 20]
new_cars = data[data['Age'] < 5]

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=[old_cars['AskPrice'], new_cars['AskPrice']], palette='Set2')
plt.xticks([0, 1], ['Old Cars (Age > 20)', 'New Cars (Age < 5)'])
plt.title('Price Distribution: Old vs. New Cars')
plt.ylabel('Asking Price')
plt.show()
```
